```python
import numpy as np
import cvxpy as cp

def calculate_dual(c, A_ub, b_ub, maximize=True):
    m, n = A_ub.shape

    # Define the primal variables
    x = cp.Variable(n)

    # Define the primal objective function coefficients
    if maximize:
        primal_c = c
    else:
        primal_c = -c

    # Define the primal constraints
    if maximize:
        primal_A_ub = A_ub
        primal_b_ub = b_ub
    else:
        primal_A_ub = -A_ub
        primal_b_ub = -b_ub

    # Define the primal problem
    primal_problem = cp.Problem(cp.Maximize(primal_c @ x), [primal_A_ub @ x <= primal_b_ub])

    # Solve the primal problem
    primal_problem.solve()

    # Get the primal solution
    primal_solution = x.value

    # Define the dual variables
    y = cp.Variable(m)

    # Define the dual objective function coefficients
    if maximize:
        dual_c = -b_ub
    else:
        dual_c = b_ub

    # Define the dual constraints
    if maximize:
        dual_A_ub = -A_ub.T
        dual_b_ub = -c
    else:
        dual_A_ub = A_ub.T
        dual_b_ub = c

    # Define the dual problem
    dual_problem = cp.Problem(cp.Minimize(dual_c @ y), [dual_A_ub @ y >= dual_b_ub])

    # Solve the dual problem
    dual_problem.solve()

    # Get the dual solution
    dual_solution = y.value

    return primal_solution, primal_problem.value, dual_solution, dual_problem.value

#1Ci
print('Question 1Ci')

# Coefficients of the objective function
c = np.array([1, 3])

# Coefficients of the inequality constraints
A_ub = np.array([[1, 1], [-1, -1], [-1, 1]])

# Right-hand side of the inequality constraints
b_ub = np.array([6, -6, 4])

# Solve the primal and dual problems
primal_sol, primal_obj, dual_sol, dual_obj = calculate_dual(c, A_ub, b_ub, maximize=True)

# Display results
print("Primal Solution:", primal_sol)
print("Primal Objective Value:", primal_obj)
print("Dual Solution:", dual_sol)
print("Dual Objective Value:", dual_obj)

#1Cii
print('Question 1Cii')
# Coefficients of the objective function
c = np.array([40, 12, 40])

# Coefficients of the inequality constraints
A_ub = np.array([[2, 1, 5], [4, 1, 1]])

# Right-hand side of the inequality constraints
b_ub = np.array([20, 30])

# Solve the primal and dual problems
primal_sol, primal_obj, dual_sol, dual_obj = calculate_dual(c, A_ub, b_ub, maximize=False)

# Display results
print("Primal Solution:", primal_sol)
print("Primal Objective Value:", primal_obj)
print("Dual Solution:", dual_sol)
print("Dual Objective Value:", dual_obj)

```
