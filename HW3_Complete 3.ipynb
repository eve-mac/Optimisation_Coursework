```python
#Q1
import cvxpy as cp
import numpy as np

def solve_primal_dual(minimization_matrix, minimization_vector):
    # Define the primal variables
    x = cp.Variable(len(minimization_vector))

    # Define the primal problem
    objective = cp.Minimize(minimization_matrix @ x + minimization_vector)
    constraints = [x >= 0]
    primal_problem = cp.Problem(objective, constraints)

    # Solve the primal problem
    primal_solution = primal_problem.solve()

    # Compute the dual problem manually
    dual_matrix = minimization_matrix.T
    dual_vector = -primal_problem.constraints[0].dual_value

    # Define the dual variables
    y = cp.Variable(len(dual_vector))

    # Define the dual problem
    dual_objective = cp.Maximize(dual_vector @ y)
    dual_constraints = [y >= 0, dual_matrix @ y <= 0]
    dual_problem = cp.Problem(dual_objective, dual_constraints)

    # Solve the dual problem
    dual_solution = dual_problem.solve()

    return primal_solution, x.value, dual_solution, y.value

# Example usage
minimization_matrix = np.array([[1, 2], [2, 3]])
minimization_vector = np.array([3, 4])

primal_solution, primal_variables, dual_solution, dual_variables = solve_primal_dual(minimization_matrix, minimization_vector)

print("Primal Solution:", primal_solution)
print("Primal Variables:", primal_variables)
print("Dual Solution:", dual_solution)
print("Dual Variables:", dual_variables)
```

```python
#Q3
from scipy.optimize import minimize

# Objective function
def objective_function(x):
    return (x[0] - 1)**2 + x[1] - 2

# Inequality constraint (g1(x) = x1 + x2 - 2)
def constraint1(x):
    return x[0] + x[1] - 2

# Equality constraint (g2(x) = -x1 + x2 - 1)
def constraint2(x):
    return -x[0] + x[1] - 1

# Initial guess
initial_guess = [0, 0]

# Define the optimization problem with constraints
constraints = [{'type': 'ineq', 'fun': constraint1}, {'type': 'eq', 'fun': constraint2}]
result = minimize(objective_function, initial_guess, constraints=constraints)

# Display the result
x_prime = result.x
lambda_prime = result.slack[0]  # Lagrange multiplier corresponding to the inequality constraint
mu_prime = result.fun  # Lagrange multiplier corresponding to the equality constraint

print("Optimal Point (x', λ', μ'):", x_prime, lambda_prime, mu_prime)
```

```python
#Q4
import numpy as np
from scipy.optimize import minimize

# Define the objective function
def objective_function(x):
    return 0.5 * ((1 - x[0])**2 + (x[1] - x[0]**2)**2)

# Initial starting points
initial_points = [np.array([0, 0]), np.array([2, 2]), np.array([-1, 1]), np.array([1, -1])]

# List to store results
results = []

# Define and solve the optimization problem for each algorithm
algorithms = ['Nelder-Mead', 'Powell', 'CG', 'BFGS']

for algorithm in algorithms:
    print(f"\nOptimizing using {algorithm} algorithm:")

    for initial_point in initial_points:
        result = minimize(objective_function, initial_point, method=algorithm)
        results.append((algorithm, initial_point, result))

        print(f"  Initial Point: {initial_point}, Iterations: {result.nit}, Minimum Value: {result.fun}")

# Analyze the impact of the initial starting point on the number of iterations
for algorithm in algorithms:
    print(f"\n{algorithm} Algorithm:")
    for initial_point in initial_points:
        iterations = [result.nit for _, point, result in results if algorithm == _ and np.array_equal(point, initial_point)]
        print(f"  Initial Point: {initial_point}, Average Iterations: {np.mean(iterations)}")
```

```python
#Q5
from scipy.optimize import linprog

#a)
# Investment opportunities
costs = [5000, 7000, 4000, 3000]  # Costs of the products
returns = [8000, 11000, 6000, 4000]  # Expected values in 1 year

# Total budget
budget = 14000

# Coefficients for the linear programming problem
c = [-return_ for return_ in returns]  # Maximize, so negate the returns
A = [[cost] for cost in costs]  # Coefficients for the inequality constraint (budget constraint)
b = [budget]  # RHS of the inequality constraint

# Bounds for the decision variables (number of each product to buy)
x_bounds = [(0, None) for _ in range(len(costs))]

# Solve the linear programming problem
result = linprog(c, A_ub=A, b_ub=b, bounds=x_bounds, method='highs')

# Extract the solution
buy_quantities = result.x

# Display the results
print("Products to Buy:")
for i, quantity in enumerate(buy_quantities):
    if quantity > 0:
        print(f"Product {chr(ord('A') + i)}: {int(quantity)} units")

print("\nTotal Portfolio Value in 1 Year:", -result.fun)

#b)
from scipy.optimize import linprog

# Investment opportunities
costs = [5000, 7000, 4000, 3000]  # Costs of the products
returns = [8000, 11000, 6000, 4000]  # Expected values in 1 year

# Total budget
budget = 14000

# Coefficients for the linear programming problem
c = [-return_ for return_ in returns]  # Maximize, so negate the returns
A_eq = [[1, 1, 1, 1]]  # Coefficients for the equality constraint (at most one unit of each product)
b_eq = [1]  # RHS of the equality constraint

# Bounds for the decision variables (number of each product to buy)
x_bounds = [(0, 1) for _ in range(len(costs))]

# Solve the linear programming problem
result = linprog(c, A_eq=A_eq, b_eq=b_eq, bounds=x_bounds, method='highs')

# Extract the solution
buy_quantities = result.x

# Display the results
print("Products to Buy:")
for i, quantity in enumerate(buy_quantities):
    if quantity > 0:
        print(f"Product {chr(ord('A') + i)}")

print("\nTotal Portfolio Value in 1 Year:", -result.fun)
```
